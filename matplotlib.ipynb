```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

1. line plot of whether and temperature

   

```python
x = [1,2,3,4,5,6,7]
y = [23,25,37,45,26,32,25]
```

```python
plt.plot(x,y,'rp--',alpha = 0.3)
plt.grid(True)
plt.title("Whether")
plt.xlabel("Days in Week")
plt.ylabel("Temperature")
plt.show()
```

It shows that the temperature is highest on wednesday and lowest in sunday. 

```python
days = [1,2,3,4,5,6,7]
max_t = [50,49,46,48,51,43,52]
min_t=[24,36,32,30,39,40,33]
avg_t=[44,39,40,41,45,44,46]
```

```python
plt.plot(days,max_t,label = "Max")
plt.plot(days,min_t,label = "Min")
plt.plot(days,avg_t,label = "Avg")
plt.xlabel("Days in week")
plt.ylabel("temperature")
plt.grid(True)
plt.legend(loc="best")
plt.show()
```

2. Bar graph vertical and horizontal of company and Revenue

   

```python
Company = ['Daraz', 'Khalti', 'Esewa','Foodmandu']
Revenue = [105,65,89,57]
Profit = [59,26,54,23]
```

```python
xpos = np.arange(len(Company))  ##changing to numpy array so we can adjust the xpos for both company and profit as below:
xpos
```

```python
plt.xticks(xpos,Company)
plt.bar(xpos-0.2,Revenue,width=0.4,label="Revenue")
plt.ylabel("Revenue")
plt.title("Nepal's Company and it's Revenue")
plt.bar(xpos+0.2,Profit,width=0.4,label="Profit")
plt.legend(loc="best")
plt.show()
```

```python
plt.yticks(xpos,Company)
plt.barh(xpos-0.1,Revenue,label="Revenue",height = 0.2)
plt.xlabel("Revenue")
plt.title("Nepal's Company and it's Revenue")
plt.barh(xpos+0.1,Profit,label="Profit",height = 0.2)
plt.legend(loc="best")
plt.show()
```

3. Pie chart  of monthly expances of me.

```python
Values = [14000, 2000, 5000,1000]
exp_labels = ['Rent','Internet bill','Food','Stationary bill']
exp_color = ['Red','Black','Yellow','Purple']
```

```python
plt.pie(Values,labels=exp_labels,colors = exp_color)
plt.legend(loc="upper right",title = "My Expenses")
plt.show()
```

4. Histogram

```python

height = [28, 23, 43, 22, 35, 44, 35, 33, 36, 13]
```

```python
plt.hist(height, bins=[10,15,30,45],rwidth = 0.95)
plt.show()
```
